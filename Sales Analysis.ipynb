```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting visual styles for the plots
sns.set(style="whitegrid")
```

```python
# Loading the data into a DataFrame
file_path = "C:/Users/USER/OneDrive - University of Lincoln/Desktop/Tracy/Work/Portofolio/Sales/Sales.csv" 
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset
print(data.head())
```

```python
# Getting a summary of the dataset
print(data.info())

# Displaying summary statistics
print(data.describe())
```

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
# Distribution of Total Sales
plt.figure(figsize=(12, 6))
sns.histplot(data['Total'], bins=30, kde=True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()

# Count plot for Customer Type
plt.figure(figsize=(12, 6))
sns.countplot(x='Customer type', data=data)
plt.title('Count of Customer Types')
plt.show()

# Count plot for Gender
plt.figure(figsize=(12, 6))
sns.countplot(x='Gender', data=data)
plt.title('Count of Customers by Gender')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Total', data=data, estimator=sum)
plt.title('Total Sales by Branch')
plt.ylabel('Total Sales')
plt.show()

avg_rating = data.groupby('Product line')['Rating'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Product line', y='Rating', data=avg_rating)
plt.title('Average Rating by Product Line')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Customer type', y='Total', data=data, estimator=sum)
plt.title('Total Sales by Customer Type')
plt.ylabel('Total Sales')
plt.show()

# Total sales by city
city_sales = data.groupby('City')['Total'].sum().reset_index().sort_values(by='Total', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Total', y='City', data=city_sales)
plt.title('Total Sales by City')
plt.ylabel('City')
plt.xlabel('Total Sales')
plt.show()
```

```python
# Total sales by city
city_sales = data.groupby('City')['Total'].sum().reset_index().sort_values(by='Total', ascending=False)
print(city_sales)

# Plot total sales by city
plt.figure(figsize=(12, 6))
sns.barplot(x='Total', y='City', data=city_sales)
plt.title('Total Sales by City')
plt.show()

# Average gross margin percentage by Product line
avg_margin = data.groupby('Product line')['gross margin percentage'].mean().reset_index()
print(avg_margin)

# Plot average gross margin percentage by Product line
plt.figure(figsize=(12, 6))
sns.barplot(x='gross margin percentage', y='Product line', data=avg_margin)
plt.title('Average Gross Margin Percentage by Product Line')
plt.show()
```
